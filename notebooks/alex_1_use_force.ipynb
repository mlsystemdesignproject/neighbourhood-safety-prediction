### 1.3 Рассмотрим более короткий датасет, с признаками IncidentDate и Borough

```python
data = df.groupby(['IncidentDate', 'Borough']).size().reset_index(name='counts')
```

```python
data.info()
```

Почему у нас число строк: 12149  - какие число уникальных районов у нас? 

```python
data[['IncidentDate', 'Borough']].groupby('IncidentDate').count()
```

```python
b1 = set(data[data['IncidentDate']=='2022-04-01']['Borough'].unique())
b2 = set(data[data['IncidentDate']=='2022-04-03']['Borough'].unique())
print(f"len(b1): {len(b1)}, len(b2): {len(b2)}")
print('b1.difference(b2)')
print(b1.difference(b2))
print('b2.difference(b1)')
print(b2.difference(b1))
print('b1.symmetric_difference(b2)')
print(b1.symmetric_difference(b2))
```
