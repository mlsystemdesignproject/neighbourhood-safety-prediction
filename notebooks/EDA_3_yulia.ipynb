# Combining MPS crime dataset with new Features

```python
crime_df = pd.read_csv('../data/MPS_agg_features/london_borough_crime.csv')
population_df = pd.read_csv('../data/MPS/population_area.csv')
attractions_df = pd.read_csv('../data/MPS/london_tourist_attractions.csv')
```
